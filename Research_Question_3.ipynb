```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# Example data
models = ['SVM', 'MNB', 'LR', 'RF', 'CNN']
accuracies_with_codesmells = [74.0, 64.0, 70.0, 72.0, 64.0]
accuracies_without_codesmells = [71.96, 73.10, 72.17, 69.49, 64.36]

# Create a dataframe for the data
df = pd.DataFrame({'Model': models * 2, 'Accuracy': accuracies_with_codesmells + accuracies_without_codesmells, 'Code Smells': ['With Code Smells'] * len(models) + ['Without Code Smells'] * len(models)})

# Create the bar plot using seaborn
sns.barplot(x='Model', y='Accuracy', hue='Code Smells', data=df)

# Set the title and labels
plt.xlabel('Model')
plt.ylabel('Accuracy')

# Show the plot
plt.show()
```

```python

# Example data
models = ['SVM', 'MNB', 'LR', 'CNN']
accuracies_with_codesmells = [69.21, 24.73, 69.21, 66.53]
accuracies_without_codesmells = [86.0, 74.0, 89.0, 15.0]

# Create a dataframe for the data
df = pd.DataFrame({'Model': models * 2, 'Accuracy': accuracies_with_codesmells + accuracies_without_codesmells, 'Code Smells': ['With Code Smells'] * len(models) + ['Without Code Smells'] * len(models)})


colors = ['#FFC300', '#FF5733']

# Create the bar plot using seaborn
ax = sns.barplot(x='Model', y='Accuracy', hue='Code Smells', data=df, dodge=True, palette=colors)

# Set the title and labels
plt.xlabel('Model')
plt.ylabel('Accuracy')

# Customize the plot
ax.grid(False)  # Remove grid lines
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border



# Show the plot
plt.show()

```
